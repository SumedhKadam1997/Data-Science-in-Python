# Which Countries have vaccinated more.

```python
Country = pd.DataFrame(list(zip(df.country.value_counts().index,df.country.value_counts())), columns=['country','count'], index=None)
```

```python
sns.factorplot(x="count", y="country",  data = Country , kind = "bar", size = 10, aspect = 1.90, palette = 'PuBuGn_d')
```

```python
df["day"] = df["date"].dt.day
df["month"] =df["date"].dt.month
df["year"] = df["date"].dt.year
```

```python
df
```

```python
fig, ax = plt.subplots(figsize = (15, 6))

sns.lineplot(df["month"], df["people_vaccinated"], hue = df["day"], palette = "mako")

ax.set_title("Seasonal people_vaccinated")
ax.set_xlabel("month")
ax.set_ylabel("people_vaccinated")
```

```python
fig, ax = plt.subplots(figsize = (15, 6))

sns.lineplot(df["month"], df["people_fully_vaccinated"], hue = df["day"], palette = "mako")

ax.set_title("Seasonal people_vaccinated")
ax.set_xlabel("month")
ax.set_ylabel("people_fully_vaccinated")
```

```python
fig, ax = plt.subplots(figsize = (15, 6))

sns.lineplot(df["month"], df["daily_vaccinations_raw"], hue = df["day"], palette = "mako")

ax.set_title("Seasonal people_vaccinated")
ax.set_xlabel("month")
ax.set_ylabel("daily_vaccinations_raw")
```

```python
ttl_people_vaccinated_per_hundred= df.sort_values(by='people_vaccinated_per_hundred', ascending=False)[:7000]
figure = plt.figure(figsize=(15,10))
sns.barplot(y=ttl_people_vaccinated_per_hundred.country, x=ttl_people_vaccinated_per_hundred.people_vaccinated_per_hundred)
plt.xticks()
plt.xlim(xmin=0.8)
plt.xlabel('people_vaccinated_per_hundred')
plt.ylabel('COUNTRY')
plt.title('Countplot of country by people_vaccinated_per_hundred')
plt.show()
```

```python
ttl_daily_vaccinations_per_million= df.sort_values(by='daily_vaccinations_per_million', ascending=False)[:7000]
figure = plt.figure(figsize=(15,10))
sns.barplot(y=ttl_daily_vaccinations_per_million.country, x=ttl_daily_vaccinations_per_million.daily_vaccinations_per_million)
plt.xticks()
plt.xlim(xmin=0.8)
plt.xlabel('daily_vaccinations_per_million')
plt.ylabel('COUNTRY')
plt.title('Countplot of country by daily_vaccinations_per_million')
plt.show()
```
